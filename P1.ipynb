## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math
from sklearn import linear_model

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def draw_lines(img, lines, color=[255, 0, 0], thickness=5):
    
    left_half_points_x = []                                              #empty array to separate left and right side lines
    left_half_points_y = []
    right_half_points_x = []
    right_half_points_y = []
    for line in lines:

        for x1, y1, x2, y2 in line:
            #check if the point sis in the first half of the image
            slope = (y2 - y1) / (x2 - x1)
            if slope != float("inf"):
                if (x1 <=img.shape[1]/2) & (x2 <=img.shape[1]/2):
                        if (slope < 0):                                  #left half line have positive slopes
                            left_half_points_x.append(x1)
                            #np.append(left_half_points_x,x1)
                            left_half_points_x.append(x2)
                            #np.append(left_half_points_x, x2)
                            left_half_points_y.append(y1)
                            #np.append(left_half_points_y, y1)
                            left_half_points_y.append(y2)
                            #np.append(left_half_points_y, y2)

                elif(x1 >= img.shape[1]/2) & (x2 >= img.shape[1]/2):    #right side lanes have negative slopes
                        if (slope > 0):  
                            right_half_points_x.append(x1)
                            #np.append(right_half_points_x, x1)
                            right_half_points_x.append(x2)
                            #np.append(right_half_points_x, x2)
                            right_half_points_y.append(y1)
                            #np.append(right_half_points_y, y1)
                            right_half_points_y.append(y2)
                            #np.append(right_half_points_y, y2)

    left_half_points_x = np.array(left_half_points_x)
    left_half_points_y = np.array(left_half_points_y)
    obj = linear_model.LinearRegression()                             #assuming x's and y's are noisy points, fit a best line using linear regression
    obj.fit(left_half_points_x[np.newaxis, :].T, left_half_points_y[np.newaxis, :].T)
    m_left = obj.coef_                                                #slopes and intercepts of left lane markings, y = m*x + b
    b_left = obj.intercept_

    right_half_points_x = np.array(right_half_points_x)
    right_half_points_y = np.array(right_half_points_y)
    obj = linear_model.LinearRegression()
    obj.fit(right_half_points_x[np.newaxis, :].T, right_half_points_y[np.newaxis, :].T)
    m_right = obj.coef_                                              #slopes and intercepts of right lane markings, y = m*x + b
    b_right = obj.intercept_

    #intersection of 2 lines, APEX
    x_intersection =int( (b_right - b_left) / (m_left - m_right))   #the right and left lanes meet at apex.
    y_intersection = int((m_left * x_intersection) + b_left)

    #left side lane bottom x coordinate
    x1 = int((img.shape[0] - b_left ) / m_left);                    #left lane lines start from x,height_of_image
    cv2.line(img, (x1, img.shape[0]), (x_intersection, y_intersection), color, thickness)

    #right bottom coordinate
    x1 = int ((img.shape[0] - b_right) / m_right);
    return cv2.line(img, (x1, img.shape[0]), (x_intersection, y_intersection), color, thickness)    
    

#not used, very trivial function!
def draw_lines_old(img, lines, color=[255, 0, 0], thickness=5):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below   """
    for line in lines:
        for x1, y1, x2, y2 in line:
            return cv2.line(img, (x1,  y1), (x2, y2), color, thickness)
    
    avg_left_points = np.empty([lines.shape[0],2])
    avg_right_points = np.empty([lines.shape[0],2])
    left_i = 0
    right_i = 0
    for line in lines:
        for x1, y1, x2, y2 in line:
            if x1 != x2: #avoiding divide by zero error
                if (x1 <=img.shape[1]/2) &  (x2 <=img.shape[1]/2):
                    avg_left_points[left_i,:] =[round((x1+x2)/2) ,round((y1+y2)/2)]
                    left_i = left_i + 1
                else:
                    avg_right_points[right_i,:] = [round((x1 + x2) / 2), round((y1 + y2) / 2)]
                    right_i = right_i + 1

    avg_left_points = avg_left_points[np.argsort(avg_left_points[0:left_i,0])]       #sort the points to get an order
    avg_right_points = avg_right_points[np.argsort(avg_right_points[0:right_i, 0])]  # sort the points to get an order

    #plot the left line
    x1 = int(avg_left_points[0,0])
    y1 = int(avg_left_points[0, 1])
    x2 = int(avg_left_points[left_i - 1, 0])
    y2 = int(avg_left_points[left_i - 1, 1])
    #drawing lines from the bottom of the video
    x_bottom = x1 - ((y1-img.shape[0])*(x2-x1)/(y2-y1))

    cv2.line(img, (int(x_bottom), img.shape[0]), (x2, y2), color, thickness)
    #plt.imshow(img)

    #for iterator in range(0,avg_left_points.shape[1]-1,2):
       # x1 = int(avg_left_points[iterator,0])
       # y1 = int(avg_left_points[iterator,1])
       # x2 = int(avg_left_points[iterator+1,0])
        #y2 = int(avg_left_points[iterator+1,1])
    x1 = int(avg_right_points[0, 0])
    y1 = int(avg_right_points[0, 1])
    x2 = int(avg_right_points[right_i-1, 0])
    y2 = int(avg_right_points[right_i-1, 1])
    # drawing lines from the bottom of the video
    x_bottom = x1 - ((y1 - img.shape[0])*(x2 - x1) / (y2 - y1))
    cv2.line(img, (int(x_bottom),  img.shape[0]), (x1, y1), color, thickness)
    #plt.imshow(img)

        #x1 = int(avg_right_points[iterator, 0])
        #y1 = int(avg_right_points[iterator, 1])
        #x2 = int(avg_right_points[iterator + 1, 0])
        #y2 = int(avg_right_points[iterator + 1, 1])
        #cv2.line(img, (x1, y1), (x2, y2), color, thickness)


def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    basic_flag = 1
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

````python
def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)
    #img = mpimg.imread(image)
    img = image
    
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # threshold yellow
    lower_yellow = np.array([20, 41, 100])
    upper_yellow = np.array([30, 255, 255])
    yellow_mask = cv2.inRange(img_hsv, lower_yellow, upper_yellow)

    lower_white = np.array([0, 0, 235])
    upper_white = np.array([255, 255, 255])
    white_mask = cv2.inRange(img_hsv, lower_white, upper_white)

    thresholded_white_yellow = cv2.bitwise_or(yellow_mask, white_mask)
    #convert to gray scale
    #grayscale_image = grayscale(img)                               #convert the image to gray scale
    kernel_size = 3                                                #define the kernel size for gaussian blurring
    grayscale_image = gaussian_blur(thresholded_white_yellow, kernel_size)  #gaussian blur the image

    #canny edge detection
    low_threshold = 50                                             #define the lower thresholding for gradient selection
    high_threshold = 150                                           #define the upper threshold for gradient selection
    edges = canny(grayscale_image, low_threshold, high_threshold)  #detect the edges using canny edge detection

    input_image_shape = edges.shape
    vertices = np.array([[(0, input_image_shape[0]), (450, 320), (500, 320), (input_image_shape[1], input_image_shape[0])]], dtype=np.int32)
    masked_image = region_of_interest(edges, vertices)

    #hough transform to find lines
    rho = 1                                                       # distance resolution in pixels of the Hough grid
    theta = np.pi / 180                                           # angular resolution in radians of the Hough grid
    threshold = 10                                                # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 10                                          # minimum number of pixels making up a line
    max_line_gap = 15                                             # maximum gap in pixels between connectable line segments
    line_image = np.copy(img) * 0                                 # creating a blank to draw lines on

    lines = hough_lines(masked_image, rho, theta, threshold, min_line_length, max_line_gap)

    line_detected_image = weighted_img(lines, img)  
    return line_detected_image
    
    
    
````
